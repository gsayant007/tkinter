```python
from PIL import Image,ImageOps
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
image = Image.open("./Lenna_(test_image).png")
plt.imshow(image)
print(image.mode)
print(image.format_description)
print(image.size)
plt.show()
```

```python
image = Image.open("./Lenna_(test_image).png")
r,g,b = image.split()
fig,ax = plt.subplots(1,3)
ax[0].imshow(r)
ax[1].imshow(g)
ax[2].imshow(b)
r.save("./Lenna_(test_image)_red.png")
g.save("./Lenna_(test_image)_green.png")
b.save("./Lenna_(test_image)_blue.png")
plt.show()
```

```python
red_arr = plt.imread("./Lenna_(test_image)_red.png")
green_arr = plt.imread("./Lenna_(test_image)_green.png")
blue_arr = plt.imread("./Lenna_(test_image)_blue.png")
print(red_arr[0,0],green_arr[0,0],blue_arr[0,0])
print(image.getpixel((0,0)))
```
